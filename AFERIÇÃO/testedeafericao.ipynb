```python
#exercicio 1-a)
def contaespaços(string):
    if string == '':
        return 0
    if string[0] == '':
        return 1 + contaespaços(string[1:])
    else:
        return contaespaços(string[1:])
```

```python
#exercicio 1 c)
def existeElem(lista, n):
    if lista == []:
        return False
    if lista[0] == n:
        return True
    return existeElem(lista[1:],n)
```

```python
#exercicio 2 a)
def maiorDivisor(N):
    dic={}
    for num in range(1, N+1):
        maior=1
        for d in range(1,10):
            if num%d == 0:
                maior=d
        dic[num] = maior
    return dic 
```

```python
#exercicio 2 b)
def mmc(a, b):
    mmc = a
    while mmc % b != 0:
        mmc += a
    return mmc
```

```python
#exercicio 2 c)
def f2c(fahrenheit):
    novo = {}
    for k in fahrenheit:
        F = fahrenheit[k]
        C = (F - 32) / 1.8
        novo[k] = (F, C)
    return novo
```

```python
#exercicio 2 d)
def reduxInt(n):
    while n >= 10:
        soma = 0
        for d in str(n):
            soma += int(d)
        n=soma
    return n
```

```python
#exercicio 3 
hospital = [["Unidade de Cardiologia"], #aqui é uma lista de [nome,medicos,camas]
    [(1, "Dr. João"), (2, "Dra. Maria")], #nos medicos sao tuplos pq nao vao mudar, sao fixos; medico=(id,nome)
    [(101, False), (102, True), (103, False), (104, False)], #assim como nas camas, que têm o numero e se estiverem ocupadas é true, livres é false
#varias camas , é uma lista de tuplos,assim como a lista dos medicos; cama=(id,ocupado)
    ["Unidade de Pediatria"],
    [(3, "Dr. Pedro"), (4, "Dra. Clara")],
    [(201, True), (202, True), (203, True)]]


 #exercicio 3a)
def listar_unidades(hospital):
    nomes=[]
    for i in range(0, len(hospital),3): #para andar de 3 em 3(onde estao as unidades); percorre a lista hospital de 3 em 3 para andar de unidade em unidade
        nomes.append(hospital[i][0]) #juntar a lista vazia de nomes o que esta no inidce 0 da lista "hospital" para cada i 
    return nomes
print(listar_unidades(hospital))
```

```python
#exercicio 3b)
def camas_disponiveis(hospital):
    res = 0 #aqui pomos res e nao criamos uma listas pq o out sera um numero, nao uma informacao 
    for i in range(2, len(hospital), 3): #aqui vemos na lista hospital que as camas aparecem no indice 2 e 5 logo comecamos em 2 e somamos mais 3 do comprimento da lista
        camas = hospital[i] #i=2:camas de cardiologia e i=5:camas de pediatria 
        for cama in camas: #agora "procuramos" na lista de camas 
            if cama[1] == False: # se o segundo elemento do tuplo cama (que é o valor booleano) for false esta livre, logo...)
                res += 1 #adicionamos mais um ao res. é o mesmo que ter res=res+1
    return res
print(camas_disponiveis(hospital))
```

```python
#exercicio 3c)
def listar_medicos_unidade(hospital,nome_unidade): #desta forma recebe 2 informacoes que queremos no argumento
    for i in range(0,len(hospital),3): #percorre apenas as unidades 
        if hospital[i][0]==nome_unidade: # se a unidade inserida corresponder à do indice 0 do hospital
            return hospital[i+1] #devolve os medicos dessa unidade,i=posicao da unidade e i+1=posicao dos medicos dessa unidade
    return [] #se nenhuma unidade tiver esse nome devolvemos uma lista vazia 
 
print(listar_medicos_unidade(hospital, "Unidade de Cardiologia")) 
```

```python
#exercicio 3d)
def existe_medico(hospital,nome_unidade,nome_medico):
    for i in range(0,len(hospital),3): #percorre as unidades que e de 3 em 3 no hospital
        if hospital[i][0]==nome_unidade:
            for medico in hospital[i+1]: #i+1 é a lista de medicos dessa unidade 
                if medico [1]==nome_medico: #cada medico e um tuplo logo medico [1] é so o nome e nao o id
                    return True 
            return False  
    return false     
print (existe_medico(hospital,"Unidade de Cardiologia","Dra. Maria"))
```

```python
#exercicio 3e) 
def adicionar_medico(hospital, nome_unidade, nome_medico):
    for i in range(0, len(hospital), 3):
        if hospital[i][0] == nome_unidade: #o i aponta sempre para o inicio de uma unidade e hospital [i] é sempre o nome de uma unidade. [i][0] da so o nome mesmo e nao a lista 
            for medico in hospital[i + 1]: # i anda de 3 em 3 e se i=0, fica hospital [1] que corresponde a lista de medicos da primeira unidade 
                if medico[1] == nome_medico:
                    return hospital
            novo_id = len(hospital[i + 1]) + 1
            hospital[i + 1].append((novo_id, nome_medico))
            return hospital
    return hospital

print(adicionar_medico(hospital, "Unidade de Cardiologia", "Dra. Paula"))
```

```python
#exercicio 3f)
def ocupar_cama(hospital, nome_unidade, id_cama):
    for i in range(len(hospital)):
        unidade = hospital[i]
        if unidade[0] == nome_unidade:
            camas = unidade[2]
            novas_camas = []
            for cama in camas:
                if cama[0] == id_cama and cama[1] == False:
                    novas_camas.append((cama[0], True))   
                else:
                    novas_camas.append(cama)
            hospital[i] = (unidade[0], unidade[1], novas_camas)
            return hospital
    return hospital

           
```

```python
#exercicio 3g) 
def guardar_hospital(hospital, nome_ficheiro):#vai receber uma string com o nome do ficheiro 
    f = open(nome_ficheiro, "w") #f é o canal para escrever no ficheiro; se o ficheiro n existir, é criado, se existir o conteudo antigo é apagado
    f.write(str(hospital)) #write so escreve strings; este str transforma a estrutura do hospital numa representacao em texto
    f.close()

```

```python
#exercicio 3h)
def carregar_hospital(nome_ficheiro):
    f = open(nome_ficheiro, "r") #abre o ficheiro recebido que ja existe e ler 
    hospital = eval(f.read())
    f.close()
    return hospital
```

```python
#exercicio 3i) 
def camas_ocupadas(hospital):
    total=0
    ocupadas:0
    for unidade in hospital: #percorremos cada unidade do hospital
        for cama in unidade[2]:#percorrer as camas dessa unidade ([2])
            total=total+1 
        if cama[1]==True:
            ocupadas=ocupadas+1
    if total==0: # se nao houver camas
        return 0
    return(ocupadas/total*100)
```

```python
#exercicio 3j)
def unidades_ocupacao_critica(hospital):
    res=[] #aqui vamos guardar so as unidades que estao em situacao >80%
    for unidade in hospital: #o mesmo codigo que na alinea anterior
        total=0
        ocupadas=0
        for cama in unidade[2]:
            total=total+1
            if cama[1]==True:
                ocupadas=ocupadas+1
        if total>0:
            perc= (ocupadas/total*100) #calculamos percentagem da unidade 
        if perc>=80: #que e o que queremos adicionar ao res (append)
            res.append((unidade[0],perc)) #criamos um tuplo e guardamos na lista res 
    return res 
        

```
